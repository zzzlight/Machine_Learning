```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the dataset
df = pd.read_csv('machine.data')
# I add the "vendor name","Model Name","MYCT","MMIN","MMAX","CACH","CHMIN","CHMAX","PRP","ERP" into dataset file before.
df.head()
df.shape
```

```python
for i in range(2,10):
    plt.figure(figsize=(6.0,4.0))
    plt.title(label=i-2)
    plt.hist(df.iloc[:,i])
    plt.show()
    plt.clf()
```

```python
feature_temp = df["MYCT"].values
PRP_list = df["PRP"].values


feature = np.array(feature_temp) # feature data
target = np.array(PRP_list) # target data

# model y = ax + b  --> get the best a and b
from sklearn.linear_model import LinearRegression
linear = LinearRegression()  
# train
linear.fit(feature.reshape(-1,1),target) 
# predict
linear.predict([[226],[333]])


x = np.linspace(0,1400,num=200)
y = linear.predict(x.reshape(-1,1))
plt.scatter(feature_temp,PRP_list)
plt.scatter(x,y)
plt.xlabel('MYCT')
plt.ylabel('PRP')
plt.title('The relation between MYCT and PRP')
```
