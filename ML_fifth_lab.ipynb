```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the dataset
train= pd.read_csv('train5.csv') #I rename the train and test to train5 and test5
test=pd.read_csv('test5.csv')

train.info()
```

```python
from sklearn.ensemble import RandomForestRegressor

age_data_train = train[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
known_age = age_data_train[age_data_train.Age.notnull()].values
unknown_age = age_data_train[age_data_train.Age.isnull()].values
y = known_age[:, 0]
X = known_age[:, 1:]
rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
rfr.fit(X, y)    
predictedAges = rfr.predict(unknown_age[:, 1::])
train.loc[ (train.Age.isnull()), 'Age' ] = predictedAges 
train.info()
train
```

```python
from sklearn import linear_model
from sklearn.ensemble import BaggingClassifier

dummies_Embarked = pd.get_dummies(train['Embarked'], prefix= 'Embarked')
train = train.join(dummies_Embarked)
dummies_Sex = pd.get_dummies(train['Sex'], prefix= 'Sex')
train = train.join(dummies_Sex)
train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model 
from sklearn import tree
from sklearn import datasets,ensemble

X_drop=['PassengerId','Name','Ticket','Cabin','Embarked','Sex','Survived']
X_train=train.drop(X_drop, axis=1).values
y_train=train['Survived'].values
#clf=ensemble.RandomForestRegressor()
clf = linear_model.LogisticRegression(penalty='l2', C=2.0,max_iter=5000) #0.79

#clf=KNeighborsClassifier(n_neighbors=5) 0.68
#clf=tree.DecisionTreeRegressor(max_depth=3)#0.37
# bagging=BaggingClassifier(clf,max_samples=0.5,max_features=0.5)
clf.fit(X_train,y_train)
```

```python
test = pd.read_csv('test5.csv')
age_data_test = test[['Age','Fare', 'Parch', 'SibSp', 'Pclass']]
unknown_age_test = age_data_test[test.Age.isnull()].values  
predictedAges_test = rfr.predict(unknown_age_test[:, 1:])
test.loc[ (test.Age.isnull()), 'Age' ] = predictedAges_test 

dummies_Embarked_test = pd.get_dummies(test['Embarked'], prefix= 'Embarked')
test = test.join(dummies_Embarked)
dummies_Sex_test = pd.get_dummies(test['Sex'], prefix= 'Sex')
test = test.join(dummies_Sex_test)
X_drop_test=['PassengerId','Name','Ticket','Cabin','Embarked','Sex']
X_test=test.drop(X_drop_test, axis=1).values
predict=clf.predict(X_test)
result = pd.DataFrame({'PassengerId':test['PassengerId'].values, 'Survived':predict.astype(np.int32)})
test = test.join(result['Survived'])
test.to_csv('test_predict.csv')
test
```

```python
from sklearn.model_selection import cross_val_score

scores=cross_val_score(clf,X_train,y_train,cv=5)

print(scores.mean())
```
